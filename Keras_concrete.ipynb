## Import headers


```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

## Build and compile model

```python
model = tf.keras.Sequential([
    keras.layers.Dense(units=10,activation='relu'),
    keras.layers.Dense(units=1,activation='relu'),
])
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

## Import data

```python
df = pd.read_csv('https://cocl.us/concrete_data')
```

```python
X = np.array([0.0 for i in range(50)])
```

## Part A: Find the average perfomance

```python
for i in range(50):
	#split data into test and train sets
	X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:7],df.iloc[:,8], test_size=0.33, random_state=42)
	
	model.fit(X_train,y_train,epochs=50)
	yhat = model.predict(X_test)
    
	#Collect the mean squared error
	X[i]  = mean_squared_error(y_test, yhat, squared=True)
```

```python
print(np.mean(X),np.std(X))
```

## Part B 

```python
for i in range(50):
    #split data into test and train sets
    X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:7],df.iloc[:,8], test_size=0.33, random_state=42)
    '''Normalizing'''
    X_train = (X_train - np.mean(X_train))/np.std(X_train)
    X_test = (X_test - np.mean(X_test))/np.std(X_test)
    y_train = (y_train - np.mean(y_train))/np.std(y_train)
    y_test = (y_test - np.mean(y_test))/np.std(y_test)
    model.fit(X_train,y_train,epochs=50)
    yhat = model.predict(X_test)
    #Collect the mean squared error
    X[i]  = mean_squared_error(y_test, yhat, squared=True)
```

```python
print(np.mean(X),np.std(X))
```
